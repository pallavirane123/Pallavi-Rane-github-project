# Calculator 

```python
import tkinter as tk
import math
```

```python
root = tk.Tk()
root.title("Arithmetic & Scientific Calculator")
root.geometry("400x450")
```

```python
label1=tk.Label(root,text="Enter No.1: ")
label2=tk.Label(root,text="Enter No.2: ")
```

```python
label1.pack(pady=5)
entry1 = tk.Entry(root, font=("Arial", 14))
entry1.pack(pady=5)

label2.pack(pady=5)
entry2 = tk.Entry(root, font=("Arial", 14))
entry2.pack(pady=5)

entry3 = tk.Entry(root, font=("Arial", 14))
entry3.pack(pady=10)
```

```python
#Arithmatic Function.
def addition():
    a=float(entry1.get())
    b=float(entry2.get())
    c=a+b
    entry3.insert(0,c)
    
def subtraction():
    a = float(entry1.get())
    b = float(entry2.get())
    c = a - b
    entry3.insert(0, c)

def multiplication():
    a = float(entry1.get())
    b = float(entry2.get())
    c = a * b
    entry3.insert(0, c)

def division():
    a = float(entry1.get())
    b = float(entry2.get())
    c = a / b
    entry3.insert(0, c)

def modulus():
    a = float(entry1.get())
    b = float(entry2.get())
    c = a % b
    entry3.insert(0, c)

# Scientific Functions.
def sine():
    a = float(entry1.get())
    c = math.sin(math.radians(a))
    entry3.insert(0, c)

def cosine():
    a = float(entry1.get())
    c = math.cos(math.radians(a))
    entry3.insert(0, c)

def tangent():
    a = float(entry1.get())
    c = math.tan(math.radians(a))
    entry3.insert(0, c)
```

```python
def delete():
    entry1.delete(0,END)
    entry2.delete(0,END)
    entry3.delete(0,END)
    
```

```python
tk.Button(root, text="Add", width=10, command=addition).pack(pady=2)
tk.Button(root, text="Subtract", width=10, command=subtraction).pack(pady=2)
tk.Button(root, text="Multiply", width=10, command=multiplication).pack(pady=2)
tk.Button(root, text="Divide", width=10, command=division).pack(pady=2)
tk.Button(root, text="Modulus", width=10, command=modulus).pack(pady=2)

tk.Button(root, text="Sin", width=10, command=sine).pack(pady=2)
tk.Button(root, text="Cos", width=10, command=cosine).pack(pady=2)
tk.Button(root, text="Tan", width=10, command=tangent).pack(pady=2)
```

```python
root.mainloop()
```
